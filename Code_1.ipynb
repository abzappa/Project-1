```python
import pandas as pd
```

```python
df1=pd.read_csv(r'customers.csv')
df1
```

```python
df2=pd.read_csv(r'order_items.csv')
df2
```

```python
df3=pd.read_csv(r'orders.csv')
df3
```

```python
df4=pd.read_csv(r'products.csv')
df4
```

```python
cust_and_ord=df1.merge(df3, how='outer', on='CustomerID')
cust_and_ord
```

```python
it_and_prod=df2.merge(df4, how='outer', on='ProductID')
it_and_prod
```

```python
merged_df=pd.merge(cust_and_ord,it_and_prod, how='outer', on='OrderID')
merged_df
```

```python
pd.set_option('display.max.columns', None)
merged_df
```

```python
merged_df=merged_df.drop(['Email','ShippingCity','ShippingCountry','OrderItemID','OrderID','ProductID'],axis=1)
merged_df
```

```python
merged_df.set_index('CustomerID', inplace=True)
merged_df
```

```python
merged_df['FullName']=merged_df['FirstName']+' '+merged_df['LastName']
merged_df
```

```python
merged_df=merged_df.drop(['FirstName','LastName'], axis=1)
merged_df
```

```python
first_column=merged_df.pop('FullName')
merged_df=pd.concat([first_column, merged_df], axis=1)
merged_df
```

```python
merged_df=merged_df.sort_index()
merged_df
```

```python
del merged_df['Price']
merged_df
```

```python
merged_df['Price']=merged_df['Quantity']*merged_df['UnitPrice']
merged_df
```

```python
col_1=merged_df.pop('Price')
merged_df.insert(9,'Price', col_1)
merged_df
```

```python
merged_df=merged_df.sort_values('OrderDate')
merged_df
```

```python
merged_df=merged_df.drop(columns=['StockQuantity'])
merged_df
```

```python
initial_mouse=150
initial_mug=300
initial_headphones=100
initial_sleeve=200
initial_lamp=75

prev_val={
    "Wireless Mouse":initial_mouse, 
    "Coffee Mug":initial_mug,
    "Bluetooth Headphones":initial_headphones, 
    "Laptop Sleeve":initial_sleeve, 
    "Desk Lamp":initial_lamp
}

merged_df['StockQuantity']=0

for i in range(len(merged_df)):
    category=merged_df.at[i, 'ProductName']
    value=merged_df.at[i, 'Quantity']

    if category=="Wireless Mouse":
        merged_df.at[i, 'StockQuantity']=prev_val[category]-value
    
    elif category=="Coffee Mug":
        merged_df.at[i, 'StockQuantity']=prev_val[category]-value
    
    elif category=="Bluetooth Headphones":
        merged_df.at[i, 'StockQuantity']=prev_val[category]-value
    
    elif category=="Laptop Sleeve":
        merged_df.at[i, 'StockQuantity']=prev_val[category]-value
    
    elif category=="Desk Lamp":
        merged_df.at[i, 'StockQuantity']=prev_val[category]-value
        
    prev_val[category]=merged_df.at[i, 'StockQuantity']

(merged_df)
```
